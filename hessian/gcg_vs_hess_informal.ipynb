```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
import torch
import torch.nn.functional as F
import transformers

import flrt.epo_v2 as epo

# The flag below controls whether to allow TF32 on matmul. This flag defaults to False
# in PyTorch 1.12 and later.
torch.backends.cuda.matmul.allow_tf32 = True

# The flag below controls whether to allow TF32 on cuDNN. This flag defaults to True.
torch.backends.cudnn.allow_tf32 = True
```

```python
def preactivation_runner(model, layer, neuron):
    def f(*model_args, **model_kwargs):
        out = {}

        def get_target(module, input, output):
            out["target"] = input[0][:, -1, neuron]

        with epo.add_fwd_hooks(
            [
                (model.model.layers[layer].mlp.activation_fn, get_target),
            ]
        ):
            out["logits"] = model(*model_args, **model_kwargs).logits
        return out

    return f
```

```python
model_name = "microsoft/phi-2"
model = transformers.AutoModelForCausalLM.from_pretrained(
    model_name,
    low_cpu_mem_usage=True,
    torch_dtype=torch.float32,
    use_cache=True,
    device_map="cuda",
)
device = model.device
for name, param in model.named_parameters():
    param.requires_grad_(False)
tokenizer = transformers.AutoTokenizer.from_pretrained(model_name)
```

```python
runner = preactivation_runner(model, layer=8, neuron=1)
```

```python
def gcg_update(pred, input_ids, topk=32, batch_size=256):
    seq_len, vocab_size = pred.shape
    position = torch.randint(0, seq_len, (batch_size,), device=device)
    token_idx = torch.randint(0, topk, (batch_size,), device=device)
    replace_ids = torch.tile(input_ids[None], (batch_size, 1)).to(device)
    ranking = (-pred).topk(k=topk, dim=-1)
    replace_ids[
        torch.arange(batch_size),
        position,
    ] = ranking.indices[position, token_idx]
    state = epo.evaluate_fitness(runner, replace_ids.reshape((-1, seq_len)), 256)
    best_target_idx = state.target.argmax(dim=-1)
    return state.ids[best_target_idx], state.target[best_target_idx], replace_ids, state
```

```python
embed = model.get_input_embeddings()
vocab_size = embed.num_embeddings
```

```python
import torch
torch.manual_seed(10000)
n_iters = 12
seq_len = 12
batch_size = 32
topk = 4
use = 'gcg'
input_ids = torch.randint(0, vocab_size, (1, seq_len,), device=device)[0]
print(tokenizer.decode(input_ids))
```

```python
for iter in range(n_iters):
    model.zero_grad()
    one_hot = F.one_hot(input_ids, num_classes=vocab_size).to(embed.weight.dtype)
    one_hot.requires_grad = True
    inputs_embeds = torch.matmul(one_hot, embed.weight)
    out = runner(inputs_embeds=inputs_embeds.unsqueeze(0))
    target = out['target']
    loss = -target
    g = torch.autograd.grad(loss.sum(), one_hot, create_graph=True)[0]

    D_expected = []
    n_samples = 1
    for i in range(n_samples):
        z = torch.randint(0, 2, g.shape, device=device) * 2.0 - 1.0
        Hz = torch.autograd.grad(g, one_hot, grad_outputs=z, retain_graph=True)[0]
        D_expected.append((Hz * z) / n_samples)
    D = sum(D_expected)
    D = torch.where(D > 0, D, torch.zeros_like(D))

    model.zero_grad()

    # f(0, 1) =~ f(1, 0) + gv + v^T H v / 2
    # v = (-1, 1)
    # H is diagonal
    # x coordinate is the current token
    # y coordinate is the proposed token
    #
    # pred1: f(0, 1) =~ f(1, 0) + (g_y - g_x)
    # pred2: f(0, 1) =~ f(1, 0) + (g_y - g_x) + (H_yy + H_xx) / 2
    pred1 = loss[:, None] + (
        g
        - g[
            torch.arange(seq_len),
            input_ids,
            None,
        ]
    )
    pred2 = (
        pred1
        + (
            D
            + D[
                torch.arange(seq_len),
                input_ids,
                None,
            ]
        )
        / 2
    )
    
    gcg_new_input_ids, gcg_target, _, _ = gcg_update(g, input_ids, topk=topk, batch_size=batch_size)
    gcg_improvement = gcg_target - out["target"]
    hess_new_input_ids, hess_target, _, _ = gcg_update(pred2, input_ids, topk=topk, batch_size=batch_size)
    hess_improvement = hess_target - out["target"]

    update_target = gcg_target if use == 'gcg' else hess_target
    new_input_ids = gcg_new_input_ids if use == 'gcg' else hess_new_input_ids
    if update_target > out["target"]:
        old_input_ids = input_ids
        input_ids = new_input_ids

    print(f"{out['target'].item():.2f} {hess_improvement.item() - gcg_improvement.item():.2f}")
```

```python
for i in range(10):
    gcg_new_input_ids, gcg_target, replace_ids, state = gcg_update(g, old_input_ids, topk=topk, batch_size=batch_size)
    old_seq = old_input_ids
    best_seq_idx = i#state.target.argmax()
    new_seq = state.ids[best_seq_idx]
    token_idx = torch.where(new_seq != old_seq)[0][0].item()
    old_token = old_seq[token_idx]
    new_token = new_seq[token_idx]
    direction_vec = torch.zeros(vocab_size, device=device)
    direction_vec[new_token] = 1
    direction_vec[old_token] = -1
    print(tokenizer.decode(old_seq))
    print(tokenizer.decode(new_seq))
    print(tokenizer.decode(old_token))
    print(tokenizer.decode(new_token))
    runner(input_ids=old_seq.unsqueeze(0))['target'], runner(input_ids=new_seq.unsqueeze(0))['target']
    one_hot_A = F.one_hot(old_seq, num_classes=vocab_size).to(embed.weight.dtype)
    one_hot_B = F.one_hot(new_seq, num_classes=vocab_size).to(embed.weight.dtype)

    beta = torch.linspace(0, 1, 100, device=model.device)[:, None, None]
    one_hot_interp = (1-beta) * one_hot_A[None, :] + beta * one_hot_B[None, :]
    one_hot_interp.requires_grad = True
    inputs_embeds = torch.matmul(one_hot_interp, embed.weight)
    out = runner(inputs_embeds=inputs_embeds)
    grads = torch.autograd.grad(out['target'].sum(), one_hot_interp)[0]
    directional_grad = (grads[:, token_idx] * direction_vec).sum(dim=-1)
    fig, ax1 = plt.subplots()
    ax2 = ax1.twinx()
    ax1.plot(beta[:, 0,0].cpu(), out['target'].detach().cpu(), 'k-')
    ax2.plot(beta[:, 0,0].cpu(), directional_grad.detach().cpu(), 'b-')
    ax2.axhline(0, color='r', linestyle='--')
    # ax2.set_ylim([-2, 2])
    plt.xlim([0, 1])
    ax1.set_xlabel('Interpolation')
    ax1.set_ylabel('Loss')
    ax2.set_ylabel('Gradient', color='b')
    plt.show()
```
